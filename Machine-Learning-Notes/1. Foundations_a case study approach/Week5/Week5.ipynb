# Collaborative filtering: People who bought this also bought...

## Co-occurrence matrix
* Matrix C: store # of users who bought both items i & j
    * \# of items x # of items
    * Symmetric $ C_{ij} = C_{ji} $
* Look at row of matrix

## Effect of popular items
* what if there are very popular items?
    * this item will be recommended for every user

## Normalizing co-occurrence matrices and leveraging purchase histories
* __Jaccard similarity__: normalize by popularity
    * Who purchased __i and j__ divided by who purchased __i or j__
* Many other similarity metrics possible, e.g., __cosine similarity__
* Limitations: 
    * only current page matters, __no history__
        - Recommend similar items to the one you bought
    * What if you purchased many items?
        - Want recommendations based on purchase history
* (Weighted) Average of purchased items
    * user bought items {A, B}
        - compute user-specific score for each item j in inventory by combining similarities:
        - $ Score(user, i) = 1/2 \times (S_{i,A} + S_{i,B}) $
        - Could also weight recent purchased more
* Limitations:
    * does not use:
        - context
        - user features
        - product features
    * Cold start problem
        - What if a new user or product arrives?

# Matrix factorization

## The matrix completion task
* Solution3: Discovering hidden structure by matrix factorization

## Recommendations from known user/item features
* Describe movie v with topics $ R_v $
* Describe user u with topics $ L_u $
* $ \hat{Rating}(u,v)$ is the product of the two vectors

## Predictions in matrix form
* $ \hat{Rating}(nxm)$ matrix is the product of vetor $L(nx1)$ and $R^{T}(1xm)$

## Discovering hidden structure by ___Matrix Factorization___
* Use observed values(ratings) to estimate "topic" vectors $\hat{L}$ and $\hat{R}$
* Use $\hat{L}$ and $\hat{R}$ for recommendations
* $ RSS(L,R) = \Sigma{(Rating(u,v) - <L_u, R_v>)^2} $  many efficient algorithms for factorization
* Limitations:
    * Cold-start problem
        - not solved yet
        
## Bring it all together: Featurized matrix factorization
* Features capture __context__
    - Time of day, what I just saw, user info, past purchases
* Discovered topics from matrix factorization capture __group of users__ who behave similarity
    - Women from Seattle who teach and have a baby

* Combine to mitigate cold-start problem
    - Ratings for a new user from features only
    - As more information about user is discovered, matrix factorization topics become more relevant
* Blending models


# Performance metrics for recommender systems
* A performane metric for recommender systems
    * Classfication accuracy = fraction of items correctly classified (liked vs. not liked)
    * Here, not interesed in what a person doesn't like
    * Rather, how quickly can we discover the relatively few likede items?
        - (partially) an imbalanced class problem
    * Recall: How many liked items were recommended
    * Precision: How many recommended items were liked
* Optimal recommenders

* Precision-recall curves
    * Input: A specific recommender system
    * Output: Algorithm-specific precision-recall curve
    * for a precision, want recall as large as possible(or vice versa)
    * One metric: largest area under the curve(AUC)
    * Another: set desired recall and maximize precision (precision at k)
